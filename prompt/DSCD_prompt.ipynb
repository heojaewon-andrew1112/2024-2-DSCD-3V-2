## 가벼운 실험 프롬프트

```python
from langchain.prompts import ChatPromptTemplate, SystemMessagePromptTemplate
from langchain.chains import LLMChain
from langchain.chat_models import ChatOpenAI

# OpenAI 연결 (gpt-3.5-turbo 모델 사용)
llm = ChatOpenAI(
    temperature=0.1,  # 창의성 (0.0 ~ 2.0)
    model_name="gpt-3.5-turbo",  # gpt-3.5-turbo 모델 사용
)

# 여행 일정 생성을 위한 템플릿
prompt_template = """
당신은 여행 일정 AI 전문가입니다. 아래의 정보를 기반으로 {city}에서 {trip_duration} 동안의 자세한 여행 일정을 생성하세요.

- 여행 날짜: {travel_dates}
- 여행 기간: {trip_duration}
- 동행인: {companions}
- 선호하는 여행 스타일: {travel_style}
- 선호하는 일정 스타일: {itinerary_style}
- 선호하는 숙소: {accommodation_preference}

**제약 조건:**
- 각 날짜마다 오전, 오후, 저녁 활동을 포함한 일정을 제공하세요.
- 각 활동에는 구체적인 관광지나 맛집을 포함하세요.
- 식당의 경우 유명한 장소나 지역의 숨은 맛집을 구체적으로 제시하세요.
- 각 장소(관광지 및 식당)의 운영시간을 포함하세요.
- 각 장소에 대해 한 문장으로 간략한 소개를 추가하세요.
- 각 장소의 정확한 주소를 포함하세요.
- ‘빼곡한 일정’을 선택한 경우 오전, 오후, 저녁마다 최소 3개의 일정을 포함하여 하루 전체에 최소 9개의 활동을 생성하세요. ‘널널한 일정’을 선택한 경우 오전, 오후, 저녁마다 최대 2개의 일정을 포함하여 하루 전체에 최대 6개의 활동을 생성하세요. 포함.
- 여행 기간 동안 동일한 장소나 식당이 중복되지 않도록 하세요.
- 동선 최적화를 위해 장소 간 이동 시간을 최소화하세요.
- 결과는 JSON 형식으로 출력하세요.
- **결과는 한국어로 제공해주세요.**
- 당신은 항상 검증 가능한 사실 진술을 바탕으로 여행 일정을 생성합니다.
- 당신은 주로 관광지 명소나 맛집에 관련된 사실적 정보 및 최근 정보에 대해 이야기하며, 자신만의 정보를 추가하지 않습니다.
"""

# 간소화된 여행 일정 생성 함수
def generate_travel_itinerary(city, trip_duration, travel_dates, companions, travel_style, itinerary_style, accommodation_preference):
    formatted_prompt = prompt_template.format(
        city=city,
        trip_duration=trip_duration,
        travel_dates=travel_dates,
        companions=companions,
        travel_style=travel_style,
        itinerary_style=itinerary_style,
        accommodation_preference=accommodation_preference
    )

    # GPT-3.5 터보 모델을 사용한 대화 생성
    response = llm.predict(formatted_prompt)
    
    return response

# 딕셔너리 형태로 일정 정보 전달
itinerary_details = {
    "city": "오사카",
    "trip_duration": "4박 5일",
    "travel_dates": "2024-10-15 ~ 2024-10-19",
    "companions": "친구와",
    "travel_style": "자연, 관광지, 맛집",
    "itinerary_style": "빼곡한 일정",
    "accommodation_preference": "5성급 호텔 이상"
}

# 여행 일정 생성 호출
itinerary = generate_travel_itinerary(**itinerary_details)

# 결과 출력
print(itinerary)
```

```python
# JSON 데이터를 파싱하여 Python 객체로 변환
data = json.loads(itinerary)

# CSV로 저장할 데이터를 담을 리스트 생성
csv_data = []

# 각 날짜별로 데이터 처리
for date, day_data in data['여행 일정'].items():
    for time_of_day, details in day_data.items():
        # CSV 데이터를 담는 리스트에 각 항목 추가
        csv_data.append([date, time_of_day, details.get('활동', details.get('식당', '')), details['운영시간'], details['소개'], details['주소']])

# Pandas DataFrame으로 변환
df = pd.DataFrame(csv_data, columns=['날짜', '시간대', '활동', '운영시간', '소개', '주소'])

df
```

## 실제 프롬프트

```python
from langchain.prompts import ChatPromptTemplate, SystemMessagePromptTemplate
from langchain.memory import ConversationSummaryBufferMemory
from langchain.chains import LLMChain
from langchain.chat_models import ChatOpenAI
from datetime import datetime

# OpenAI 연결
llm = ChatOpenAI(
    temperature=0.1,  # 창의성 (0.0 ~ 2.0)
    model_name="gpt-4o",  # GPT-4 모델
)

# 여행 일정 생성을 위한 템플릿
prompt_template = """
당신은 여행 일정 AI 전문가입니다. 아래의 정보를 기반으로 {city}에서 {trip_duration} 동안의 자세한 여행 일정을 생성하세요.

- 여행 날짜: {travel_dates}
- 여행 기간: {trip_duration}
- 동행인: {companions}
- 선호하는 여행 스타일: {travel_style}
- 선호하는 일정 스타일: {itinerary_style}
- 선호하는 숙소: {accommodation_preference}

**제약 조건:**
- 각 날짜마다 오전, 오후, 저녁 활동을 포함한 일정을 제공하세요.
- 각 활동에는 구체적인 관광지나 맛집을 포함하세요.
- 식당의 경우 유명한 장소나 지역의 숨은 맛집을 구체적으로 제시하세요.
- 각 장소(관광지 및 식당)의 운영시간을 포함하세요.
- 각 장소에 대해 한 문장으로 간략한 소개를 추가하세요.
- 각 장소의 정확한 주소를 포함하세요.
- 선호하는 일정 스타일로 ‘빼곡한 일정’을 선택한 경우 하루에 최소 11개의 활동을 포함하세요. ‘널널한 일정’을 선택한 경우 하루에 최대 7개의 활동을 포함하세요.
- 여행 기간 동안 동일한 장소나 식당이 중복되지 않도록 하세요.
- 동선 최적화를 위해 장소 간 이동 시간을 최소화하세요.
- 결과는 JSON 형식으로 출력하세요.
- **결과는 한국어로 제공해주세요.**

"""

# 페르소나 설정
persona = """
You are a travel AI expert specialized in generating optimized travel itineraries.

You always generate travel itineraries based on verifiable factual statements.
You mainly speak about factual information related to tourist attractions and restaurants, and recent updates, without adding information on your own.
"""
```

```python
from langchain.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate, MessagesPlaceholder
from langchain.memory import ConversationSummaryBufferMemory
from langchain.chains import LLMChain
from langchain.chat_models import ChatOpenAI


# 여행 일정 생성 함수
def generate_travel_itinerary(city, trip_duration, travel_dates, companions, travel_style, itinerary_style, accommodation_preference):
    # 페르소나 주입
    filled_persona = persona.format()
    
    formatted_prompt = prompt_template.format(
        city=itinerary_details['city'],
        trip_duration=itinerary_details['trip_duration'],
        travel_dates=itinerary_details['travel_dates'],
        companions=itinerary_details['companions'],
        travel_style=itinerary_details['travel_style'],
        itinerary_style=itinerary_details['itinerary_style'],
        accommodation_preference=itinerary_details['accommodation_preference']
    )


    # 프롬프트 구성
    prompt = ChatPromptTemplate(
        template=formatted_prompt,
        messages=[
            SystemMessagePromptTemplate.from_template(filled_persona),  # 페르소나 주입
            MessagesPlaceholder(variable_name="chat_history"),  # 메모리 활용
            HumanMessagePromptTemplate.from_template("{question}")
        ]
    )

    # 메모리 설정
    memory = ConversationSummaryBufferMemory(llm=llm, max_token_limit=1000, memory_key="chat_history", return_messages=True)

    # LLMChain 설정
    conversation = LLMChain(
        llm=llm,
        prompt=prompt,
        verbose=True,
        memory=memory
    )

    # 여러 입력 변수를 하나의 딕셔너리로 전달
    result = conversation({
        "question": formatted_prompt
    })
    
    return result['text']


# 딕셔너리 형태로 일정 정보 전달
itinerary_details = {
    "city": "프랑스 파리",
    "trip_duration": "3박 4일",
    "travel_dates": "2024-11-15 ~ 2024-11-18",
    "companions": "친구와",
    "travel_style": "쇼핑, 관광지, 맛집",
    "itinerary_style": "빼곡한 일정",
    "accommodation_preference": "5성급 호텔 이상"
}

# 여행 일정 생성 호출
itinerary = generate_travel_itinerary(**itinerary_details)

# 결과 출력
print(itinerary)
```

```python
import json
import pandas as pd

# 불필요한 설명 문장 및 뒤쪽 텍스트 제거하고 JSON 부분만 추출
start_index = itinerary.find("{")  # JSON 시작 위치 찾기
end_index = itinerary.rfind("}")   # JSON 끝 위치 찾기

# JSON 부분만 추출
json_text = itinerary[start_index:end_index+1]

# JSON 문자열을 파싱하여 Python 객체로 변환
data = json.loads(json_text)

# CSV로 저장하기 위한 리스트 생성
csv_data = []

# 날짜별로 데이터를 처리
for day in data['여행 일정']:
    date = day['날짜']
    
    # 각 시간대(오전, 오후 등)별로 데이터 처리
    for time_of_day in ['오전', '오후', '저녁']:
        if time_of_day in day:
            for activity in day[time_of_day]:
                csv_data.append([
                    date,               # 날짜
                    time_of_day,        # 시간대
                    activity['활동'],    # 활동
                    activity['소개'],    # 소개
                    activity['운영시간'], # 운영시간
                    activity['주소']     # 주소
                ])

# Pandas DataFrame으로 변환
df = pd.DataFrame(csv_data, columns=['날짜', '시간대', '활동', '소개', '운영시간', '주소'])

df
```
